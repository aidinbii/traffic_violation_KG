```python
def generate_horizbar(df_1, title, xlabel, filename=None, save=False):
    fig, ax_1 = plt.subplots(figsize=(18,6))
    plt.barh(list(df_1.index),df_1.values,)
    ax_1.set_title(title)
    ax_1.set_xlabel(xlabel)
    for i in ax_1.patches:
        percentage = '{:.2f}%'.format(100*i.get_width() / sum(df_1.values))
        y = i.get_y()+i.get_height()-0.5
        x = i.get_width()
        ax_1.annotate(percentage, (x, y))
    if save:
        fig.savefig(f'./saved_figures/{filename}.png', dpi=600,bbox_inches='tight')
```

```python
def generate_vertbar_fine(df_1, title, xlabel, ylabel, filename=None, save=False):
    '''
    This function generates a vertical bar plot specifically for fine values.
    '''
    fig, ax_1 = plt.subplots(figsize=(10,6))
    x_val = np.array(df_1.index)
    count = np.array(df_1.values)
    y_val = x_val * count
    print (y_val)
    ax=sns.barplot(x = x_val, y = y_val,width=0.5)
    ax_1.set_title(title)
    ax_1.set_xlabel(xlabel)
    ax_1.set_ylabel(ylabel)
    if save:
        fig.savefig(f'./saved_figures/{filename}.png', dpi=300,bbox_inches='tight')
```

```python
df_location = df[df['crossroad_name'].isin(location_list)]
```

```python
df_location.groupby(['crossroad_name', 'car' ])['id'].count().sort_values()
```

```python
dict_location = dict(list(df_location.groupby(['crossroad_name'])))
```

```python
# sort according to location_list order
dict_location = dict(sorted(dict_location.items(), key=lambda pair: location_list.index(pair[0])))
```

```python
dict_location;
```

```python
top_cars_cross = []
for loc, data in dict_location.items():
    print (loc)
    # taking only top-5 cars in each of the top-5 crossroads
    print (dict(data.groupby(['car']).count()['id'].sort_values(ascending=False)[:5]))
    top_cars_cross.append(dict(data.groupby(['car']).count()['id'].sort_values(ascending=False)[:5]))
```

```python
# sanity check
sum(top_cars_cross[0].values())
```

```python
top_cars_cross
```

```python
# Convert numpy.int64 to int in dictionary
# https://stackoverflow.com/questions/50916422/python-typeerror-object-of-type-int64-is-not-json-serializable
for c in top_cars_cross:
    for key in c.keys():
        c[key] = int(c[key])
```

```python
type((top_cars_cross[0]['HONDA FIT']))
```

```python
# convert to str
import json
top_cars_str = []
for c in top_cars_cross:
    car_str = json.dumps(c)
    car_str = car_str[1:-1]
    # Remove the quotation marks from the JSON string
    car_str = car_str.replace('"', '')
    top_cars_str.append(car_str)
```

```python
top_cars_str
```

```python
df_crossroad['top_cars_cross'] = top_cars_str
```

```python
# Convert data to GeoJSON format
features = []
for index, row in df_crossroad.iterrows():
    feature = {
        'type': 'Feature',
        'geometry': {
            'type': 'Point',
            'coordinates': [row['longitude'], row['latitude']]
        },
        
        'properties': {
            'Name': row['crossroad_name'],
            'Top-5 cars': row['top_cars_cross'].replace(',', ',<br>')
            #'Description': row['Description'],
            #'Category': row['Category']
        }
    }
    features.append(feature)

geojson_data = {
    'type': 'FeatureCollection',
    'features': features
}
```

# Filter by year

```python
import july
from july.utils import date_range
from sklearn.preprocessing import normalize
```

```python
def plot_heatmap_by_year(year, month_1, month_2, normalize=False, save=False):
    ''' 
    Plot heatmap for every year
    month_1 - the start month
    month_2 - the end month
    for a given year
    '''
    df_year = df[df['violation_date'].dt.strftime('%Y') == year]
    count_day = df_year.groupby([df_year['violation_date'].dt.month, df_year['violation_date'].dt.day]).size()
    count_day_values = count_day.values
    
    # normalize the values
    if normalize:
        count_day_values = count_day_values / np.linalg.norm(count_day_values)
        
    july.heatmap(date_range(f"{year}-{month_1}-01", f"{year}-{month_2}-31"),count_day_values , 
                 cmap="pastel_sunrise", colorbar=True, month_grid=True, title="Number of violations")
    if save:
        plt.savefig(f"./saved_figures/heatmap_by_year_{year}.png")
```

```python
plot_heatmap_by_year('2019','09','12', normalize=False, save=True)
```

```python
def plot_bar_traffic_by_year_hour(year,save=False):
    ''' 
    Plot number of traffic-violations in 24 hours for every year
    '''
    df_year = df[df['violation_date'].dt.strftime('%Y') == year]
    
    df_4 = (df_year.groupby([df_year['violation_date'].dt.hour,])['id'].count())
    
    fig, ax_1 = plt.subplots(figsize=(10,6))
    x_val = np.array(df_4.index)
    y_val = np.array(df_4.values)
    ax=sns.barplot(x = x_val, y = y_val,width=0.5)
    ax_1.set_title(f'Number of traffic-violations in 24 hours, year {year}')
    ax_1.set_xlabel('Hours')
    ax_1.set_ylabel('Number of violations')
    
    if save:
        plt.savefig(f"./saved_figures/viol_24h_by_year_{year}.png")
        
    return fig, ax_1
```

```python
fig_1, ax_1 = plot_bar_traffic_by_year_hour('2019', save=True)
```

```python
fig_2, ax_2 = plot_bar_traffic_by_year_hour('2020', save=True)
```

```python
fig_3, ax_3 = plot_bar_traffic_by_year_hour('2021', save=True)
```

## combine four bar plots into one 

```python
def get_values_24hours(year):
    df_year = df[df['violation_date'].dt.strftime('%Y') == year]
    df_4 = (df_year.groupby([df_year['violation_date'].dt.hour,])['id'].count())
    x_val = np.array(df_4.index)
    y_val = np.array(df_4.values)
    return x_val, y_val
```

```python
x_1,y_1 = get_values_24hours('2019')
x_2,y_2 = get_values_24hours('2020')
x_3,y_3 = get_values_24hours('2021')
```

```python
# for all years
df_5 = (df.groupby([df['violation_date'].dt.hour,])['id'].count())
x_all = np.array(df_5.index)
y_all = np.array(df_5.values)
```

```python
width = 0.3
fontsize = 10
fig, ax = plt.subplots(nrows=2, ncols=2, figsize=(10, 8))

fig.suptitle('Number of traffic violations in 24 hours', fontsize=12)

sns.barplot(x=x_all, y=y_all,ax=ax[0, 0],width=width)
ax[0, 0].set_title('Whole period', fontsize=fontsize)
#ax[0, 0].set_xlabel(xlabel)
ax[0, 0].set_ylabel(ylabel)

sns.barplot(x=x_1, y=y_1,ax=ax[0, 1],width=width)
ax[0, 1].set_title('Year 2019', fontsize=fontsize)
#ax[0, 1].set_xlabel(xlabel)
#ax[0, 1].set_ylabel(ylabel)

sns.barplot(x=x_2, y=y_2,ax=ax[1, 0],width=width)
ax[1, 0].set_title('Year 2020', fontsize=fontsize)
ax[1, 0].set_xlabel(xlabel)
ax[1, 0].set_ylabel(ylabel)

sns.barplot(x = x_3, y = y_3, ax=ax[1, 1], width=width)
ax[1, 1].set_title('Year 2021', fontsize=fontsize)
ax[1, 1].set_xlabel(xlabel)
#ax[1, 1].set_ylabel(ylabel)

plt.tight_layout()
plt.show()
```
